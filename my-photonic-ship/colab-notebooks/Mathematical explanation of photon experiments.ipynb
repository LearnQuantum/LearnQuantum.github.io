## Basic states

Like a classical bit, a quantum bit (qubit) has two basis states. However, we use "ket" notation to represent them:
- $\ket{0}$ reprents the state 0 and
- $\ket{1}$ reprents the state 1.


## NOT gate (operator)

It flips the value:
- $X \ket{0} = \ket{1} $ and
- $X \ket{1} = \ket{0} $.

## Hadamard gate (operator)

It creates a combination of the basis state:

- $ H \ket{0} = \dfrac{1}{\sqrt{2}} \ket{0} + \dfrac{1}{\sqrt{2}} \ket{1} $

and

- $ H \ket{1} = \dfrac{1}{\sqrt{2}} \ket{0} - \dfrac{1}{\sqrt{2}} \ket{1} $.

Here the scalar values are called as _amplitudes_.

If measured, we observe the states $\ket{0}$ and $ \ket{1} $ with probability $\dfrac{1}{2}$ after applying a Hadamard to a basis state.



## Amplitudes and probabilities

The probabilities are calculated based on amplitudes.

For example, a qubit can be in a quantum state as

$ \ket{u} = a \ket{0} + b \ket{1} $.

Then, the probabilities of being in states $\ket{0}$ and $\ket{1}$ are $ |a|^2 $ and $|b|^2$.

As the total probability must be 1, then we have

$ |a|^2 + |b|^2 = 1 $.

_Note that $ a $ and $b$ can also be complex numbers._

## Quantum superposition

In the above example, if both $a$ and $b$ are non-zero, then we say that the qubit is in a superposition (of $\ket{0}$ and $\ket{1}$).

When in a superposition, applying a quantum operator (e.g., Hadamard) can create interferences.

## Re-visting the photon experiments

### Experiment 1

We start in state $\ket{0}$.

We apply a Hadamard operator. Then, our quantum state is

$ \ket{+} = \dfrac{1}{\sqrt{2}} \ket{0} + \dfrac{1}{\sqrt{2}} \ket{1} $.

We make a measurement.

Thus, we obtain $\ket{0}$ and $\ket{1}$ with probability $ \dfrac{1}{2} $.

After the measurement, the state of qubit is the observed outcome.

### Experiment 2

We repeat Experiment 1, but we continue in the case of $ \ket{1} $.

We have $\ket{0}$ and $\ket{1}$ with probability $ \dfrac{1}{2} $ after the measurement.

If the measurement outcome is $\ket{1}$, we apply a second Hadamard. Our quantum state in this branch is

$ \ket{-} = \dfrac{1}{\sqrt{2}} \ket{0} - \dfrac{1}{\sqrt{2}} \ket{1} $

After measuring, we obtain $\ket{0}$ and $\ket{1}$ with probability $ \dfrac{1}{2} $.

Thus, we can observe $\ket{1}$ with probability $ \dfrac{1}{2} \cdot \dfrac{1}{2} = \dfrac{1}{4} $.

And, we can observe $ \ket{0} $ with probability $ \dfrac{1}{2} + \dfrac{1}{4} = \dfrac{3}{4}  $.


### Experiment 3

We start in $\ket{0}$ and apply a Hadamard operator twice.

After the first Hadamard:

$ H \ket{0} = \ket{+} = \dfrac{1}{\sqrt{2}} \ket{0} + \dfrac{1}{\sqrt{2}} \ket{1} $.

After the second Hadamard:

$ H \ket{+} = H \left( \dfrac{1}{\sqrt{2}} \ket{0} + \dfrac{1}{\sqrt{2}} \ket{1} \right) = \dfrac{1}{\sqrt{2}} H \ket{0} + \dfrac{1}{\sqrt{2}} H \ket{1} $.

We obtain four outcomes and we add them all:

$ \dfrac{1}{\sqrt{2}} \left(
  \left( \underbrace{\dfrac{1}{\sqrt{2}} \ket{0} + \dfrac{1}{\sqrt{2}} \ket{1}}_{H\ket{0}} \right)
  +
  \left( \underbrace{ \dfrac{1}{\sqrt{2}} \ket{0} - \dfrac{1}{\sqrt{2}} \ket{1} }_{H \ket{1}} \right)  \right) $.

  That is:
  $ \left( \dfrac{1}{2} \ket{0} + \dfrac{1}{2} \ket{0} \right)
   +
   \left( \dfrac{1}{2} \ket{1} - \dfrac{1}{2} \ket{1} \right)  = \ket{0} $.

Here $\ket{0}$s are interfered constructively, and $\ket{1}$s are interfered destructively.

## $ Z $ Gate

It flips the sign of $ \ket{1} $.

If we are in $ \ket{+} $ and apply $Z$, we obtain $ \ket{-} $:

$ Z \left(  \dfrac{1}{\sqrt{2}} \ket{0} + \dfrac{1}{\sqrt{2}} \ket{1} \right)
= \dfrac{1}{\sqrt{2}} \ket{0} - \dfrac{1}{\sqrt{2}} \ket{1} = \ket{-} $.

Similarly, if we are in $ \ket{-} $ and apply $Z$, we obtain $ \ket{+} $.

Using Matrices and Vectors

This part is optional.

We represent the quantum states as vectors.

- $ \ket{0} = \vzero $ and

- $ \ket{1} = \vone $.

We represent the operators (gates) as matrices:

$ I = \I,~  X = \X, ~ H = \hadamard , ~ Z = \mymatrix{cr}{1 & 0 \\ 0 & -1} $

We observe that

$ HH = XX = ZZ = I $.

Applying a quantum operator, say $U$ to a quantum state, say $\ket{u}$, is represented as amtrix-vector multiplication:

$ \ket{v} =  U \ket{u} $ , where $ \ket{v} $ is the new quantum state.
